```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
t1 = pd.read_csv('Mortality/T1.csv', )
t1.columns = t1.iloc[0]
t1 = t1[1:]
t1.reset_index(drop=True, inplace=True)
```

```python
t1.columns = t1.columns.astype(str)
t1.columns = [col.strip('0') for col in t1.columns]
t1.columns = [col.strip('.') for col in t1.columns]
t1.columns
```

```python
t1.head()
```

```python
t1.shape
```

```python
t1.drop(['Indicator Code', 'Country Code'], axis=1, inplace=True)
```

```python
australia = t1.loc[t1['Country Name']=='Australia']
australia
```

```python
australia = australia.transpose()
australia.columns = australia.iloc[0]
australia = australia[2:]
```

```python
australia
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(australia['Australia'])
plt.xticks(rotation=90)

plt.show()
```
